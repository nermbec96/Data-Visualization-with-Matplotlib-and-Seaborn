# Data Visualization with Matplotlib and Seaborn

#### Visualizing data can help us to understend data and do better analysis. In this project I will do data visulization in Python using Matplotlib and Seaborn. The data I am using are crypto, daily prices and returns, and breast cancer dataset

```python
#Importing pcakages 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import datasets
```

```python
#Importnig datasets 

daily_return_df = pd.read_csv(r'/Users/nerminbecirovic/Desktop/project 3 /crypto_daily_returns.csv') 
daily_return_df
```

```python
investments_df = pd.read_csv(r'/Users/nerminbecirovic/Desktop/project 3 /crypto_daily_prices.csv') 
investments_df
```

```python
cancer = datasets.load_breast_cancer()
cancer_df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
cancer_df
```

In first part of this project I will do visualization using Seaborn. Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. For this part of project is will use breast cancer dataset.


##### A scatter plot is used to observe and visually display the relationship between variables, in this example I will use a scatterplot to present relationship between the mean area and the mean smoothness.

```python
plt.figure(figsize = (8,8))
sns.scatterplot(x = 'mean area', y = 'mean smoothness',hue = 'target',data = cancer_df)
plt.title('Relationship between mean area and mean smoothness')
plt.show()
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = 'mean radius',y = 'mean area',hue ='target',data= cancer_df)
plt.title('Relationship between mean radius and mean area')
plt.show()
```

##### Rigplot is the method used to plot data and a linear regression model fit

```python
 sns.regplot(x = 'mean smoothness',y = 'mean compactness',data= cancer_df)
```

##### Pairplot, heatmaps, displot are really heplful methods to use becuse it giving me many insights about new data especialiy if is classification type problem. Using this methods we can uderstend data much better.

```python
sns.pairplot(cancer_df,hue = 'target',vars = ['mean radius','mean texture','mean area','mean smoothness'])
```

```python
plt.figure(figsize = (30,30))
sns.heatmap(cancer_df.corr(),annot = True)
```

```python
plt.figure(figsize = (7,7))
sns.displot(cancer_df['mean radius'],bins = 25, color = 'orange',kde = True)
```

In the secound part of this project I will do visualization using Matplotlib. Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is also a really helpful method I usually use for my analysis.

##### Here is expamle of lineplots. We can have one or more variables in one plot and also we can make subplots.

```python
investments_df.plot(x = 'Date', y = 'BTC-USD Price', label = 'Bitcoin Price', figsize = (15, 10), linewidth = 3,color = 'blue')
plt.ylabel('Price')
plt.title('Bitcoin Price')
plt.legend(loc = "upper left")
plt.grid()
```

```python
investments_df.plot(x = 'Date', y = ['BTC-USD Price', 'ETH-USD Price', 'LTC-USD Price'], label = ['Bitcoin Price', 'Etheremun Prace', 'LiteCoin Prace'])
plt.ylabel('Price [$]')
plt.title('Crypto Prices')
plt.legend(loc = "upper center")
plt.grid()
```

```python
investments_df.plot(x = 'Date', title = 'Crypto Prices', subplots = True, grid = True, figsize = (15, 25))
#Also other way to make subplots
#plt.figure(figsize = (15,25))
#plt.subplot(3,1,1)
#plt.plot(investments_df['BTC-USD Price'],'r-')
#plt.grid()
#plt.subplot(3,1,2)
#plt.plot(investments_df['ETH-USD Price'],'b--')
#plt.grid()
#plt.subplot(3,1,3)
#plt.plot(investments_df['LTC-USD Price'],'g-.')
#plt.grid()
```

```python
plt.figure(figsize = (25,10))
plt.subplot(1,2,1) 
plt.plot(investments_df['BTC-USD Price'],'r-') 
plt.grid()
plt.title('BTC-USD Price')
plt.subplot(1,2,2)
plt.plot(investments_df['ETH-USD Price'],'b--')
plt.title('ETH-USD Price')
plt.grid()
```

##### Here is example of histagram method. A histogram is basically used to represent data provided in a form of some groups.

```python
mu = daily_return_df['ETH'].mean()
sigma = daily_return_df['ETH'].std()
num_bins = 40
plt.figure(figsize = (20, 15))
daily_return_df['ETH'].plot.hist(bins = 30, alpha = 0.5, facecolor = 'g', figsize = (15, 8)) 
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu) + ', sigma=' +str(sigma))
plt.grid()
```

##### Here is example of pie chart. A Pie Chart is a circular statistical plot that can display only one series of data

```python
my_dict = {'allocation %': [10, 20, 15, 35, 10]}
explode = (0.1, 0, 0, 0.1, 0)
crypto_df = pd.DataFrame(data = my_dict, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA']) 
crypto_df.plot.pie(y = 'allocation %',explode = explode, figsize = (8, 8))
plt.title('CRYPTO PORTFOLIO PIE CHART')
```

##### 3D plot example

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection = '3d')
x = daily_return_df['BTC'].tolist() 
y = daily_return_df['ETH'].tolist()
z = daily_return_df['LTC'].tolist()
ax.scatter(x, y, z, c = 'b', marker = 'o', s = 400)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```
